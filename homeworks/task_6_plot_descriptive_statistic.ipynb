## Task NÂª 6: Plot histogram of DIC and TA.

```python
import pandas as pd

def read_csv_file_pandas(file_path):
    """
    Reads a CSV file using pandas and returns a DataFrame.
    
    :param file_path: Path to the CSV file
    :return: pandas DataFrame
    """
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None
    except Exception as e:
        print(f"Error: {e}")
        return None
```

### CHARGE DATAFRAME

```python
csv_df = read_csv_file_pandas("E:\Folders\Desktop\My_Things\Modelab2025_Course\course_python_rep\Modelab-Python\data\Terminos_lagoon_TA_DIC_2023_RawData.csv")
if csv_df is not None:
    print(csv_df)
```

### CHARGE MY PACKAGES


```python
import matplotlib.pyplot as plt

# PLOT DIC
dic_df = csv_df["dic_micromol_kg"]
plt.hist(dic_df, bins=25)
plt.xlabel("dic_micromol_kg")
plt.ylabel("Frecuency")
plt.show()

# PLOT TA
ta_df = csv_df["ta_micromol_kg"]
plt.hist(ta_df, bins=25)
plt.xlabel("ta_micromol_kg")
plt.ylabel("Frecuency")
plt.show()
```
